```python
#@title Card Class
class Card:
    # Represents a standard game card
    def __init__(self, suit = 0, value = 2, cardback = 0):
        self.suit = suit
        self.value = value
        self.cardback = cardback
    
    suit_names = ['Paus', 'Ouros', 'Copas', 'Espadas']
    value_names = [None, 'Ás', '2', '3', '4', '5', '6', '7', '8', '9', '10',
                    'Valete', 'Dama', 'Rei']
    cardback_names = ['Preto', 'Vermelho']
    
    def __str__(self):
        return '%s de %s de fundo %s' % (Card.value_names[self.value],
                             Card.suit_names[self.suit],
                             Card.cardback_names[self.cardback])
    def __lt__(self, other):
        t1 = self.suit, self.value, self.cardback
        t2 = other.suit, other.value, other.cardback
        return t1 < t2
```

```python
#@title Deck Class
import random
class Deck:
    '''Represents a standard deck'''
    def __init__(self):
        self.cards = []
        for cardback in range(2):
            for suit in range(4):
                for value in range(1, 14):
                    card = Card(suit, value, cardback)
                    self.cards.append(card)
    def __str__(self):
        cards = []
        for card in self.cards:
            cards.append(display_card(card))
        return '\n'.join(cards)

    def display(self):
        ''' Displays each card in hand '''
        for card in self.cards:
            print(card)

    def pop_card(self):
        return self.cards.pop()
    
    def add_card(self, card):
        self.cards.append(card)
    
    def shuffle(self):
        random.shuffle(self.cards)
    
    def move_cards(self, hand, num):
        for i in range(num):
            hand.add_card(self.pop_card())
```

```python
#@title Hand Class
class Hand(Deck):
    ''' Represents the player's hand'''
    def __init__(self, label = ''):
        self.cards = []
        self.label = label
    
    def order(self):
        self.cards.sort()
    
    def order_cardvalue(self):
        self.cards.sort(key=lambda card: card.value)
    
    def order_reverse_cardvalue(self):
        self.cards.sort(key=lambda card: card.value, reverse=True)
    
    def display_card_at_index(hand, index):
        if (index >= 0 and index < len(hand.cards)):
            #print(f'A carta no index {index} é {hand.cards[index]}')
            return hand.cards[index]
        else:
            print('Index fora de escopo')

    def remove_card_by_index(self, index):
        if (index >= 0 and index < len(self.cards)):
            removed_card = self.cards.pop(index)
            print(f'{removed_card} foi removida da sua mão')
        else:
            print('Index fora de escopo. Nenhuma carta foi removida da mão')

    def check_hand_length(hand):
        return len(hand.cards)

    def add_card_to_hand(hand, card):
        hand.cards.append(card)
        print(f'{card} foi adicionado a sua mão')
```

```python
#@title Function that evaluates which type of hand the player has
def evaluate_hand(hand):
    values = [card.value for card in hand.cards]
    if 1 in values and 10 in values and 11 in values and 12 in values and 13 in values:
        values[values.index(1)] += 13
    values.sort(reverse = True)
    straight = True
    for i in range(1, 5):
        if values[i] != values[i-1] - 1:
            straight = False
    flush = True
    for i in range(1, 5):
        if hand.cards[i].suit != hand.cards[i-1].suit:
            flush = False
    if values[0] == 14 and values[1] == 13 and values[2] == 12 and values[3] == 11 and values[4] == 10 and flush:
        return 'Royal Flush'
    elif values[0] == values[1] and values[1] == values[2] and values[2] == values[3] and values[3] == values[4]:
        return 'Five of a Kind'
    elif straight and flush:
        return 'Straight Flush'
    elif values[0] == values[1] and values[1] == values[2] and values[2] == values[3] or values[1] == values[2] and values[2] == values[3] and values[3] == values[4]:
        return 'Four of a Kind'
    elif values[0] == values[1] and values[2] == values[3] and values[3] == values[4] or values[0] == values[1] and values[1] == values[2] and values[3] == values[4]:
        return 'Full House'
    elif flush:
        return 'Flush'
    elif straight:
        return 'Straight'
    elif values[0] == values[1] and values[1] == values[2] or values[1] == values[2] and values[2] == values[3] or values[2] == values[3] and values[3] == values[4]:
        return 'Three of a Kind'
    elif values[0] == values[1] and values[3] == values[4] or values[1] == values[2] and values[3] == values[4] or values[0] == values[1] and values[2] == values[3] or values[0] == values[1] and values[2] == values[4] or values[1] == values[2] and values[3] == values[4]:
        return 'Two Pairs'
    elif values[0] == values[1] or values[1] == values[2] or values[2] == values[3] or values[3] == values[4]:
        return 'Pair'
    else:
        return 'High Card'
```

```python
#@title Function that attributes a num value to a hand type for different hand type comparisons
def quantify_hand(evaluate_hand):
    hand_value: int = 0
    if evaluate_hand == 'Royal Flush':
        hand_value = 11 
    elif evaluate_hand == 'Five of a Kind':
        hand_value = 10
    elif evaluate_hand == 'Straight Flush':
        hand_value = 9
    elif evaluate_hand == 'Four of a Kind':
        hand_value = 8
    elif evaluate_hand == 'Full House':
        hand_value = 7
    elif evaluate_hand == 'Flush':
        hand_value = 6
    elif evaluate_hand == 'Straight':
        hand_value = 5
    elif evaluate_hand == 'Three of a Kind':
        hand_value = 4
    elif evaluate_hand == 'Two Pairs':
        hand_value = 3
    elif evaluate_hand == 'Pair':
        hand_value = 2
    elif evaluate_hand == 'High Card':
        hand_value = 1

    return hand_value
```

```python
#@title Function that compares two hands and returns 0 (if a tie), 1 (if hand one is > hand two) and 2 (if hand one is < than hand two)
def compare_hands(hand1, hand2):
    
    #we reverse order the hands for high cards comparisons
    hand1.order_reverse_cardvalue()
    hand2.order_reverse_cardvalue()

     #ininitalize list with each card value type por comparisons
    values1:list = [card.value for card in hand1.cards]
    values2:list = [card.value for card in hand2.cards]
    suits1:list = [card.suit for card in hand1.cards]
    suits2:list = [card.suit for card in hand2.cards]
    cardbacks1:list = [card.cardback for card in hand1.cards]
    cardbacks2:list = [card.cardback for card in hand2.cards]

    ##################################
    ######### Value criteria #########
    ##################################

     #if its 'four of a kind'
    if evaluate_hand(hand1) == 'Four of a Kind' or evaluate_hand(hand1) == 'Full House' or evaluate_hand(hand1) == 'Three of a Kind':
         #the middle card in the hand will always be part of the four/three, since the hands are ordered, so we compare hands using it
        if values1[2] > values2[2]:
            print(f'\nA primeira mão tem o {evaluate_hand(hand1)} mais alto')
            return 1
        elif values2[2] > values1[2]:
            print(f'\nA segunda mão tem o {evaluate_hand(hand2)} mais alto')
            return 2
    
     #if its 'two pairs'
    if evaluate_hand(hand1) == 'Two Pairs':
         #Since the hands are ordered its logically necessary for one of the cards
         #of the first pair to be at index 1 and one of the cards of the second pair to be at index 3
         #checks first pair
        if values1[1] > values2[1]:
            print('\nA primeira mão tem o par mais alto')
            return 1
        elif values2[1] > values1[1]:
            print('\nA segunda mão tem o par mais alto')
            return 2
          #checks second pair
        elif values1[3] > values2[3]:
            print('\nA primeira mão tem o segundo par mais alto')
            return 1
        elif values2[3] > values1[3]:
            print('\nA segunda mão tem o segundo par mais alto')
            return 2

     #if its a 'pair'
    if evaluate_hand(hand1) == 'Pair':
         #Since the cards are ordered its logically necessary for one of the cards of the pair to be at index 1, 2 or 3
        pair1: int = 0
        pair2: int = 0
        for i in range(4):
            if values1[i] == values1[i + 1]:
                pair1 = values1[i]
                break
        for i in range(4):
            if values1[i] == values1[i + 1]:
                pair2 = values2[i]
                break 
        if pair1 > pair2:
            print(f'\nA primeira mão tem o {evaluate_hand(hand1)} mais alto')
            return 1
        if pair2 > pair1:
            print(f'\nA segunda mão tem o {evaluate_hand(hand2)} mais alto')
            return 2

     #if its a 'high card' in itself or for untiying reasons'            
    for i in range(len(values1)):
        if values1[i] > values2[i]:
            print('\nA primeira mão tem a carta mais alta')
            return 1
        elif values2[i] > values1[i]:
            print('\nA segunda mão tem a carta mais alta')
            return 2
    print('\nAs duas mãos tem os mesmos valores de cartas')

    #################################
    ######### Suit criteria #########
    #################################

    #if its 'four of a kind', 'full house' or 'three of a kind'
    if evaluate_hand(hand1) == 'Four of a Kind' or evaluate_hand(hand1) == 'Full House' or evaluate_hand(hand1) == 'Three of a Kind':
        fours1: list = []
        fours2: list = []
        for i in range(5):
            if values1[i] == values1[3]:
                fours1.append(suits1[i])
        for i in range(5):
            if values2[i] == values2[3]:
                fours2.append(suits2[i])
        fours1.sort(reverse=True)
        fours2.sort(reverse=True)
        for i in range(len(fours1)):
            if fours1[i] > fours2[i]:
                print(f'\nA primeira mão tem o maior naipe entre o {evaluate_hand(hand1)}')
                return 1
            elif fours2[i] > fours1[i]:
                print(f'\nA segunda mão tem o maior naipe entre o {evaluate_hand(hand2)}')
                return 2
    
     #if its 'two pairs'
    if evaluate_hand(hand1) == 'Two Pairs':
        pair1_1: list = []
        pair2_1: list = []
        pair1_2: list = []
        pair2_2: list = []
        for i in range(5): 
            if values1[i] == values1[1]:
                pair1_1.append(suits1[i])
            elif values1[i] == values1[3]:
                pair1_2.append(suits1[i])
        for i in range(5): 
            if values2[i] == values2[1]:
                pair2_1.append(suits2[i])
            elif values2[i] == values2[3]:
                pair2_2.append(suits2[i])
        pair1_1.sort(reverse=True)
        pair2_1.sort(reverse=True)
        for i in range(2):
            if pair1_1 > pair2_1:
                print('\nA primeira mão tem o maior naipe entre o primeiro par.')
                return 1
            elif pair2_1 > pair1_1:
                print('\nA segunda mão tem o maior naipe entre o primeiro par.')
                return 2
        pair1_2.sort(reverse=True)
        pair2_2.sort(reverse=True)
        for i in range(2):
            if pair1_2 > pair2_2:
                print('\nA primeira mão tem o maior naipe entre o segundo par')
                return 1
            elif pair2_2 > pair1_2:
                print('\nA segunda mão tem o maior naipe entre o segundo par')
                return 2

     #if its a 'pair'
    if evaluate_hand(hand1) == 'Pair':
        pair1_list: list = []
        pair2_list: list = []
        for i in range(5):
            if values1[i] == pair1:
                pair1_list.append(suits1[i])
        for i in range(5):
            if values1[i] == pair2:
                pair2_list.append(suits2[i]) 
        if pair1_list > pair2_list:
            print('\nA primeira mão tem o maior naipe entre o par')
            return 1
        if pair2_list > pair1_list:
            print('\nA segunda mão tem o maior naipe entre o par')
            return 2

      #if its a 'high card' in itself or for untiying reasons'          
    suits1.sort(reverse=True)
    suits2.sort(reverse=True)
    for i in range(len(values1)):
        if suits1[i] > suits2[i]:
            print('\nA primeira mão tem o maior naipe entre as cartas iguais')
            return 1
        elif suits2[i] > suits1[i]:
            print('\nA segunda mão tem o maior naipe entre as cartas iguais')
            return 2
    print('\nAs duas mãos tem os mesmos naipes entre as cartas iguais')

    #################################
    ####### Cardback criteria #######
    #################################

    #if its 'four of a kind', 'full house' or 'three of a kind'
    if evaluate_hand(hand1) == 'Four of a Kind' or evaluate_hand(hand1) == 'Full House' or evaluate_hand(hand1) == 'Three of a Kind':
        fours1: list = []
        fours2: list = []
        for i in range(5):
            if values1[i] == values1[3]:
                fours1.append(cardbacks1[i])
        for i in range(5):
            if values2[i] == values2[3]:
                fours2.append(cardbacks2[i])
        fours1.sort(reverse=True)
        fours2.sort(reverse=True)
        for i in range(len(fours1)):
            if fours1[i] > fours2[i]:
                print(f'\nA primeira mão tem uma cor de baralho maior entre o {evaluate_hand(hand1)}')
                return 1
            elif fours2[i] > fours1[i]:
                print(f'\nA segunda mão tem uma cor de baralho maior entre o {evaluate_hand(hand2)}')
                return 2
    
     #if its 'two pairs'
    if evaluate_hand(hand1) == 'Two Pairs':
        pair1_1: list = []
        pair2_1: list = []
        pair1_2: list = []
        pair2_2: list = []
        for i in range(5): 
            if values1[i] == values1[1]:
                pair1_1.append(cardbacks1[i])
            elif values1[i] == values1[3]:
                pair1_2.append(cardbacks1[i])
        for i in range(5): 
            if values2[i] == values2[1]:
                pair2_1.append(cardbacks2[i])
            elif values2[i] == values2[3]:
                pair2_2.append(cardbacks2[i])
        pair1_1.sort(reverse=True)
        pair2_1.sort(reverse=True)
        for i in range(2):
            if pair1_1 > pair2_1:
                print('\nA primeira mão tem uma cor de baralho maior entre o primeiro par')
                return 1
            elif pair2_1 > pair1_1:
                print('\nA segunda mão tem uma cor de baralho maior entre o primeiro par')
                return 2
        pair1_2.sort(reverse=True)
        pair2_2.sort(reverse=True)
        for i in range(2):
            if pair1_2 > pair2_2:
                print('\nA primeira mão tem uma cor de baralho maior entre o segundo par')
                return 1
            elif pair2_2 > pair1_2:
                print('\nA segunda mão tem uma cor de baralho maior entre o segundo par')
                return 2

     #if its a 'pair'
    if evaluate_hand(hand1) == 'Pair':
        pair1_list: list = []
        pair2_list: list = []
        for i in range(5):
            if values1[i] == pair1:
                pair1_list.append(cardbacks1[i])
        for i in range(5):
            if values1[i] == pair2:
                pair2_list.append(cardbacks2[i]) 
        if pair1_list > pair2_list:
            print(f'\nA primeira mão tem uma cor de baralho maior entre o {evaluate_hand(hand1)}')
            return 1
        if pair2_list > pair1_list:
            print(f'\nA segunda mão tem uma cor de baralho maior entre o {evaluate_hand(hand2)}')
            return 2

      #if its a 'high card' in itself or for untiying reasons'          
    cardbacks1.sort(reverse=True)
    cardbacks2.sort(reverse=True)
    for i in range(len(values1)):
        if cardbacks1[i] > cardbacks2[i]:
            print('\nA primeira mão tem uma cor de baralho maior entre as cartas iguais')
            return 1
        elif cardbacks2[i] > cardbacks1[i]:
            print('\nA segunda mão tem uma cor de baralho maior entre as cartas iguais')
            return 2
    print('\nAs duas mão tema as mesmas cores de baralho entre as cartas iguais')
    return 0 
```

```python
#@title Function that executes player. The cards in hand need to be previously assigned
from google.colab import output
def player(hand):
            #deck.move_cards(hand)
            hand.order()

            print(f'{hand.label} \n{hand}\n')
            #hand1.display()

            while len(hand.cards) > 5:
                discard = False
                while discard is False:  
                    card_remove = int(input('\nQue carta você quer remover da mão? \n(Forneça o número da posição da carta na sua mão)\n'))
                    certain = input(f'\nVocê tem certeza de que quer remover: {hand.display_card_at_index(card_remove - 1)}? \n(sim/nao): ').lower()
                    if certain == 'sim':
                        output.clear()
                        discard = True
                    elif certain == 'nao':
                        print('\nTenha certeza que escolheu a carta certa na próxima vez')
                    else:
                        print('\nEssa não é uma resposta válida!')
                print('\n')
                hand.remove_card_by_index(card_remove - 1)
                print('\nSua mão agora é: ')
                #hand1.display()
                print(f'\n{hand}')
            if hand.check_hand_length() == 5: 
                print(f'\n{evaluate_hand(hand)}')
            
            donelooking: bool = False
            while donelooking is False:
                looked = input('\nTerminou de olhar sua mão final? \n(sim/nao)').lower()
                if looked == 'sim':
                    donelooking = True
                elif looked == 'nao':
                    print('Por favor, termine de olhar a sua mão')
                else:
                    print('Essa não é uma resposta válida!')
```

```python
#@title Poker Game
from google.colab import output

#variable to verify if more than one game will be played
play_again: bool = True

#loops while game is active
while play_again is True:

        #variables that count the amount of times a player won a round
        player1_wins: int = 0
        player2_wins: int = 0

        #variable that tells which round it is
        current_round: int = 0

        #function that displays the rules before the game
        rules()
        
        #loops while no player has won the game
        while player1_wins < 3 and player2_wins < 3:
            ###########################################
            ################# Start! ##################
            ###########################################
            
            #initializes deck
            deck = Deck()
            deck.shuffle()

            current_round += 1

            #displays the amount of time each player won a round
            print(f'\nRound: {current_round}')
            if current_round > 1:
                print(f'\nO primeiro jogador já ganhou {player1_wins} rounds')
                print(f'\nO segundo jogador já ganhou {player2_wins} rounds')

            #variable that verifies if the  round should begin
            begin: bool = False

            while begin is False:
                new_round: str = input('\nPodemos começar o round? \n(sim/nao)').lower()
                if new_round == 'sim':
                    begin = True
                elif new_round == 'nao':
                    print('\nComeçaremos o round quando estiver pronto')
                else:
                    print('\nEscreva sim para começarmos o round!')
            output.clear()

            ###########################################
            ################ PLAYER: 1 ################
            ###########################################

            #defines player hand
            hand1 = Hand('Jogador 1')
            #deals cards to the player
            deck.move_cards(hand1, 7)
            #executes player
            player(hand1)
            output.clear()

            ###########################################
            ################ PLAYER: 2 ################
            ###########################################
            
            #defines player hand
            hand2 = Hand('Jogador 2')
            #deals cards to the player
            deck.move_cards(hand2, 7)
            #executes player
            player(hand2)
            output.clear()

            #displays the type of each hand before comparing them
            print(f'\nA mão do primeiro jogador é: {evaluate_hand(hand1)}')
            print(f'\nA mão do segundo jogador é: {evaluate_hand(hand2)}')

            #compares both hands for type
            if quantify_hand(evaluate_hand(hand1)) > quantify_hand(evaluate_hand(hand2)):
                print('\nO primeiro jogador ganhou o round!')
                player1_wins += 1
            elif quantify_hand(evaluate_hand(hand1)) < quantify_hand(evaluate_hand(hand2)):
                print('\nO segundo jogador ganhou o round!')
                player2_wins += 1
              #if the comparison between hand types is a tie compares the other criteria
            elif quantify_hand(evaluate_hand(hand1)) == quantify_hand(evaluate_hand(hand2)):

                #variable that stores which player won or tied
                winner: int = compare_hands(hand1, hand2)

                #if player 1 won
                if winner == 1:
                    print('\nO primeiro jogador ganhou o round!')
                    player1_wins += 1
                #if player 2 won
                elif winner == 2:
                    print('\nO segundo jogador ganhou o round!')
                    player2_wins += 1
                # if its still a tie
                elif winner == 0:

                    #variable that stores if the game remains a tie
                    tie: bool = True
                    
                    #counts how many times the game has tied for the current round
                    tied_times: int = 1

                    #loops while 1 cards is dealt to each player and it remains a tie
                    while tie == True:
                        #display how many times the game has tied
                        print(f'\nO número de vezes que os jogadores já empataram neste round foi: {tied_times}')
                        
                        #initializes deck
                        deck = Deck()
                        deck.shuffle()

                        ###########################################
                        ################ PLAYER: 1 ################
                        ###########################################
                        
                        #defines player hand
                        hand1 = Hand('Jogador 1')
                        #deals cards to the player
                        deck.move_cards(hand1, 1)
                        #executes player
                        player(hand1)

                        ###########################################
                        ################ PLAYER: 2 ################
                        ###########################################
                        
                        #defines player hand
                        hand2 = Hand('Jogador 1')
                        #deals cards to the player
                        deck.move_cards(hand2, 1)
                        #executes player
                        player(hand2)

                        if winner == 1:
                            print('\nO primeiro jogador ganhou o round!')
                            player1_wins += 1
                            tie = False
                        elif winner == 2:
                            print('\nO segundo jogador ganhou o round!')
                            player2_wins += 1
                            tie = False
                        tied_times += 1

        #checks how many times each player won
        #if player one won ends current game
        if player1_wins == 3:
            print('\nO primeiro jogador ganhou o jogo!')
        #if player two won ends current game
        elif player2_wins == 3:
            print('\nO segundo jogador ganhou o jogo!')

        #checks if player want to play the game again    
        while 0 == 0:
                restart: str = input('\nJogar de novo? (sim/nao)')
                if restart == 'sim':
                    print('Reiniciando jogo...')
                elif restart == 'nao':
                    print('Fim de jogo!')
                    play_again = False
                    break
                else:
                    print('\nEssa não é uma resposta válida!')       
```
