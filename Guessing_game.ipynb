```python
print('+----------------------------------+')
print('| Bem-vindo ao jogo LEEGA apostas! |')
print('+----------------------------------+')

import random

while 0 == 0:
  mode: str = input('\nEscolha o modo de jogo:\n1.Adivinhe a letra!\n2.Adivinhe um número de 1 até 200!\nEscolha: ')
  if mode == '1':
    while 0 == 0:
      dificulty: str = input('\nEscolha a dificuldade:\n1.Fácil - 15 tentativas \n2.Médio - 10 tentativas \n3.Difícil - 5 tentativas\nEscolha: ')
      if dificulty == "1":
        tentativas: int = 15
        print(f'\nFácil: Você tem {tentativas} tentativas para descobrir a letra secreta...')
        break
      elif dificulty == "2":
          tentativas: str = 10
          print(f'\nMédio: Você tem {tentativas} tentativas para descobrir a letra secreta...')
          break
      elif dificulty == "3":
          tentativas:str = 5
          print(f'\nDifícil: Você tem {tentativas} tentativas para descobrir a letra secreta...')
          break
      else:
        print("Valor inválido!")

    alphabet: str = 'abcdefghijklmnopqrstuvwxyz'
    secret_letter: str = random.choice(alphabet) #Seleciona uma letra aleatória do alfabeto (lower case)
    index_sl: int = alphabet.index(secret_letter) #Pega o valor númerico da letra aleatória na string para comparação
    pontuacao: int = 1000
    
    for rodada in range(1, tentativas + 1):
        print(f'\nRODADA {rodada}')
        guess = input("Qual é o seu palpite?\nDigite uma letra entre a e z: ").lower()

    #Alerta o usuário para entrada inválida quando digitado um numérico
        if guess.isdecimal():
            print("\nDigite apenas letras!\n")
            print(f'Tente novamente!\nTentativas restantes: {tentativas}')
            continue
    #Alerta o usuário para entrada inválida quando digitado mais de um character   
        if len(guess) != 1:
            print('\nDigite apenas uma letra!')
            print(f'Tente novamente!\nTentativas restantes: {tentativas}')
            continue
        try:
            index_guess: int = alphabet.index(guess) #Pega o valor numérico da letra chutada na string para comparação
        except ValueError as ve:
            print("\nCaracteres especiais não são válidos!")
            print(f'Tente novamente!\nTentativas restantes: {tentativas}')
            continue

        if secret_letter == guess:
           print(f'\nParabéns! Você acertou, a letra secreta é: {secret_letter}')
           print(f'Pontuação: {pontuacao}')
           break
        else:
            if index_sl > index_guess:
                print(f'\nSeu palpite {guess} é MENOR que a letra secreta.')
            elif index_sl < index_guess:
                print(f'\nSeu palpite {guess} é MAIOR que a letra secreta.')

            tentativas = tentativas - 1
            pontuacao = pontuacao - (abs(index_sl - index_guess))

            if tentativas > 0:

                print(f'Tente novamente!\nTentativas restantes: {tentativas}')
            else:
                print('\n################')
                print('# Fim de Jogo! #')
                print('################')
                print(f'A letra secreta era {secret_letter}')
                print(f'Sua pontuação: {pontuacao}')
    
    if tentativas == 0 and pontuacao == 1000:
      print('\n################')
      print('# Fim de Jogo! #')
      print('################')
      print(f'O número secreto era {secret_letter}')
      print('Sua pontuação: 0')
      break
    
  if mode == '2':
    while 0 == 0:
      dificulty: str = input("\nEscolha sua dificuldade:\n1.Hyper Fácil - 25 tentativas\n2.Fácil - 20 tentativas\n3.Médio - 15 tentativas\n4.Difícil - 10 tentativas\n5.LEEGADO - 5 tentativas\nEscolha: ")
      if dificulty == "1":
        tentativas = 25
        print(f'\nHyper Fácil: Você tem {tentativas} tentativas para descobrir o número secreto...')
        break
      elif dificulty == "2":
        tentativas: int = 20
        print(f'\nFácil: Você tem {tentativas} tentativas para descobrir o número secreto...')
        break
      elif dificulty == "3":
        tentativas: int = 15
        print(f'\nMédio: Você tem {tentativas} tentativas para descobrir o número secreto...')
        break
      elif dificulty == "4":
        tentativas: int = 10
        print(f'\nDifícil: Você tem {tentativas} tentativas para descobrir o número secreto...')
        break
      elif dificulty == "5":
        tentativas: int = 5
        print(f'\nLEEGADO: Você tem {tentativas} tentativas para descobrir o número secreto...')
        break
      else:
        print("Valor inválido!")

    secret_number: int = random.randint(1, 200) 
    pontuacao: int = 1000
    
    for rodada in range(1, tentativas + 1):
        print(f'\nRODADA {rodada}')
        guess = input("Qual é o seu palpite?\nDigite um número entre 1 e 200: ")
        
        if guess.isdecimal():
            guess = int(guess)

            # Palpite fora do intervalo
            if guess < 1 or guess > 200:
              print("Fora do Intervalo!\n")
              if tentativas > 0:
                  print(f'Tente novamente!\nTentativas restantes: {tentativas}')
              continue
        else:
          print("\nDigite apenas valores inteiros!")
          print(f'Tente novamente!\nTentativas restantes: {tentativas}')
          continue

        if secret_number == guess:
          print(f'\nParabéns! Você acertou, o número secreto é: {secret_number}')
          print(f'Pontuação: {pontuacao}')
          break
        else:
            if secret_number > guess: #palpite menor
                print(f'\nSeu palpite {guess} é MENOR que o número secreto.')
            elif secret_number < guess: #palpite maior
                print(f'\nSeu palpite {guess} é MAIOR que o número secreto.')

            tentativas = tentativas - 1
            pontuacao = pontuacao - (abs(secret_number - guess))

            if tentativas > 0:
                print(f'Tente novamente!\nTentativas restantes: {tentativas}')
            else:
                print('\n################')
                print('# Fim de Jogo! #')
                print('################')
                print(f'O número secreto era {secret_number}')
                print(f'Sua pontuação: {pontuacao}')

  if tentativas == 0 and pontuacao == 1000:
      print('\n################')
      print('# Fim de Jogo! #')
      print('################')
      print(f'O número secreto era {secret_number}')
      print('Sua pontuação: 0')

  else:
      print("\nValor inválido!")
      print('Por favor escolha entre um dos modos de jogo')
```
