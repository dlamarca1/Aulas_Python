```python
import getpass
import random

def get_secret_word():
  while True:
    secret_word: str = getpass.getpass().lower()
    if secret_word.isalpha() and secret_word.isascii():
      break
    else: 
      print('Por favor digite apenas palavras')
  return secret_word

def get_name():
  while True:
    name: str = input().lower()
    if name.isalpha() and name.isascii():
      break
    else: 
      print('Por favor digite apenas palavras')
  return name

def display_hangman(tries):
    stages = [  '''
                    +----+
                     |   |
                     O   |
                    /|\  |
                     |   |
                    / \  |
                         |
                    =========
                ''',
                '''
                    +----+
                     |   |
                     O   |
                    /|\  |
                     |   |
                    /    |
                         |
                    =========
                ''',
                '''
                    +----+
                     |   |
                     O   |
                    /|\  |
                     |   |
                         |
                         |
                    =========
                ''',
                '''
                    +----+
                     |   |
                     O   |
                    /|\  |
                         |
                         |
                         |
                    =========
                ''',
                '''
                    +----+
                     |   |
                     O   |
                    /|   |
                         |
                         |
                         |
                    =========
                ''',
                '''
                    +----+
                     |   |
                     O   |
                     |   |
                         |
                         |
                         |
                    =========
                ''',
                '''
                    +----+
                     |   |
                     O   |
                         |
                         |
                         |
                         |
                    =========
                ''',
                '''
                    +----+
                     |   |
                         |
                         |
                         |
                         |
                         |
                    =========
                '''
    ]
    return stages[tries]

print('+-----------------------------+')
print('| Bem-vindo ao jogo da forca! |')
print('+-----------------------------+')

print('\nEste jogo envolve 6 pessoas: \n5 Jogadores sorteados aleatoriamente e um informante da palavra a ser adivinhada')
print('Antes de começar é necessário informar quem são os 5 jogadores, a palavra a ser adivinhada e uma dica')
print('Não são aceitas palavras ou nomes com caracteres especiais, incluindo espaço')

print('\nNomeie os 5 jogadores: ')
players: list = []

number_player: int = 0
while number_player < 5:

    print(f'\nNome {number_player + 1}: ')
    verify_name = get_name()
    if verify_name not in players:
        players.append(verify_name)
        number_player += 1
    else:
        print('Nome já está na lista!')

round: int = 0
while round < 5:

    round += 1

    print(f'\nDigite a {round}ª palavra: ')
    word: str = get_secret_word()

    print('\nDigite uma dica: ')
    hint: str = get_secret_word()

    while hint == word:
        print('A dica não pode ser igual a palavra!')
        print('\nDigite uma nova dica: ')
        hint: str = get_secret_word()

    word_completion: str = '_' * len(word)
    guessed_letters: list = []
    guessed_words: list = []
    tries: int = 7
    guessed: bool = False

    print('\n')
    print('+------------------+')
    print('| Round:', round, 'Start!  |')
    print('+------------------+')

    random.shuffle(players)
    print(f'\nO {round}º jogador é: {players.pop(0)}')

    while not guessed and tries > 0:
    
        print(f'Você ainda tem: {tries} tentativas')
        print(f'Você já tentou as seguintes letras: {guessed_letters}')
        print(f'Você já tentou as seguintes palavras: {guessed_words}')
        print(display_hangman(tries))
        print(f'{word_completion}    O tamanho da palavra é: {len(word)}')
        print(f'A dica é: {hint}')

        guess: str = input('\nPor favor tente adivinhar uma letra ou a palavra: ').lower()
        if len(guess) == 1 and guess.isalpha() and guess.isascii():
            if guess in guessed_letters:
                print('\nVocê já tentou essa letra!', guess)
            elif guess not in word:
                print(f'\n{guess} não faz parte da palavra')
                tries -= 1
                guessed_letters.append(guess)
            else:
                print(f'\n{guess} faz parte da palavra!')
                guessed_letters.append(guess)
                word_as_list: list = list(word_completion)
                indices = [i for i, letter in enumerate(word) if letter == guess]
                for index in indices:
                    word_as_list[index] = guess
                word_completion = ''.join(word_as_list)
                if '_' not in word_completion:
                    guessed = True
        elif len(guess) == len(word) and guess.isalpha() and guess.isascii():
            if guess in guessed_words:
                print('\nVocê já tentou essa palavra!')
            elif guess != word:
                print(guess, '\nnão é a palavra correta!')
                tries -= 1
                guessed_words.append(guess)
            else:
                guessed = True
                word_completion = word
    
    
        else:
            print('\nEsse palpite não é válido')

        print('\n')

    if guessed:
        print('Parabéns, você acertou a palavra!')
    else:
        print('Infelizmente acabaram suas tentativas :(')
        print(f'A palavra era: {word}')
```
